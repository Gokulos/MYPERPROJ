## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## PRE-PROCESSING

```python
df=pd.read_csv(r"dataset_phishing.csv")
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
Y = df.loc[:, ['status']]
Y
```

```python
X=df.drop(['status','url'],axis=1)
X
```

```python
df.dtypes
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)


y_train_fixed = np.ravel( Y_train) 
y_train_fixed
```

## Random Forest 

```python
model = RandomForestClassifier()
model.fit(X_train,y_train_fixed)
```

```python
y_pred = model.predict(X_test)
```

```python
c = confusion_matrix(Y_test, y_pred)
print(f"confusion_matriix: \n {c}")

print("\n Classification Score:::::::::::::::::::::::::::::::: ")

print(classification_report(Y_test, y_pred))

print("\n Accuracy Sore")

accuracy = accuracy_score(Y_test, y_pred)

print(f" Accuracy: {accuracy}")
```
